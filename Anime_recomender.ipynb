```python
recc=anime.copy()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 13500,stop_words="english")
```

```python
cv.fit_transform(recc["Tags"]).toarray()
```

```python
cv.fit_transform(recc["Tags"]).toarray().shape
```

```python
vectors=cv.fit_transform(recc["Tags"]).toarray()
```

```python
cv.get_feature_names_out()[0:10]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
def Recommender(user_input):
    Index_of_anime = recc[recc["Name"] == user_input].index[0]
    Similarity_score = similarity[Index_of_anime]
    Sorted_scores = sorted(list(enumerate(Similarity_score)),reverse = True,key= lambda x: x[1]) [1:6]

    for i in Sorted_scores:
        print(recc.iloc[i[0]].Name)
```

```python
Recommender("Mo Dao Zu Shi 3")
```
